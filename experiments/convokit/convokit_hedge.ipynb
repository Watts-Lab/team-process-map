```python
import pandas as pd
from convokit import PolitenessStrategies, Corpus, download
from convokit import User, Utterance,Speaker
import os

#REPLACE THIS WITH YOUR OWN COMPUTER'S PATH
os.environ['REQUESTS_CA_BUNDLE'] = '/Library/Frameworks/Python.framework/Versions/3.10/lib/python3.10/site-packages/certifi/cacert.pem'
```

```python

# Create a default speaker
default_speaker = Speaker(id="user", name="User") #EXPERIMENT!!

data = pd.read_csv('jury_TINY_output_chat_level.csv')
data['message'].fillna('', inplace=True)
data['speaker_nickname'].fillna('', inplace=True)

# Create a default speaker
default_speaker = Speaker(id="user", name="User")

# Convert the extracted columns into lists
message_list = data['message'].tolist()
speaker_nickname_list = data['speaker_nickname'].tolist()
print("here")

id = 0
utterances = []
for text, nickname in zip(message_list, speaker_nickname_list):
    if text is not None and nickname is not None:
        speaker = Speaker(id=nickname, name=nickname)
        utterance = Utterance(id = str(id),conversation_id = str(0),text=text,speaker=speaker)
        utterances.append(utterance)
        print(utterance)
        id = id+1

# Create a Corpus object and add the Utterance objects
corpus = Corpus(utterances=utterances)
```

```python
ps = PolitenessStrategies(verbose=100)
corpus = ps.transform(corpus)
```

```python
values = []
idx = []
for utterance in corpus.iter_utterances():
    values.append(utterance.meta["politeness_strategies"])
    idx.append(utterance.id)
pd.DataFrame(values, index=idx).to_csv("results.csv")
```
